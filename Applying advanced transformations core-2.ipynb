```python
hei_wei['Weight'] = hei_wei['Weight'].str.replace('kg',"")

## Apply the json.loads to the full column
hei_wei['Weight'] = hei_wei['Weight'].apply(json.loads)
hei_wei['Weight'].head()
```

## Combining both data frames

```python
super_hero_df = pd.merge(df3,df1, left_on= 'Hero', right_on= 'hero_names', how = 'inner')
display(super_hero_df.info())
display(super_hero_df.head())
```

```python
# moving Hero column to first column on left side
herocol = super_hero_df.pop('Hero')
super_hero_df.insert(0,'Hero',herocol)

# deleting heros_name column from combined df3
super_hero_df.drop(columns='hero_names')
```

# 1- Compare the average weight of super powers who have Super Speed to those who do not.

```python
average_height_superspeed = super_hero_df.groupby('Super Speed')['Weight'].mean().round(2)
average_height_superspeed
```

-Average weight of super speed heros is 129.40 kilo

-Average weight of non super speed heros is 101.77 kilo

-Super speed heros on average weigh more than non super speed heros
